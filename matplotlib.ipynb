# matplotlib workflow


```python
#1 import
x=[1,2,3,4]
y=[11,22,33,44]
#2 setup plot
fig,ax=plt.subplots(figsize=(10,10)) #width ,height
# plot data
ax.plot(x,y)
#5 customize the plot
ax.set(title="simple plot",
        xlabel="X-axis",
        ylabel="Y-axis")
#save the plot
fig.savefig("plot.png")
```

# making figures using array
we want 
*line plot
*scatter plot
*bar plot
*histogram
*subpoints

```python
#creating some data
x=np.linspace(0,10,100)
x[:20]
```

```python
#plot the data and acreate a line plot
fig,ax=plt.subplots()
ax.plot(x,x**2)
```

```python
#scatter plot
fig,ax=plt.subplots()
ax.scatter(x,np.sin(x));
```

```python
#bar plot
butter_prices={"amul butter":30,
              "saras butter" :25,
              "urmul butter":23}
fig,ax=plt.subplots()
ax.bar(butter_prices.keys(),butter_prices.values())
ax.set(title="Butter Prices",
        xlabel="Compines",
        ylabel="Price(Rs)")
```

```python
#horizontal bar Graph
fig,ax=plt.subplots()
ax.barh(butter_prices.keys(),butter_prices.values())
```

```python
# make Histogram
W=np.random.randn(100)
fig,ax=plt.subplots()
ax.hist(W);
```

```python
#Subplots
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,
                                      ncols=2,
                                      figsize=(10,5))
ax1.plot(x,x**2)
ax2.scatter(x,np.sin(x))
ax3.hist(W);
ax4.bar(butter_prices.keys(),butter_prices.values())
```
